```python
# Import Dependencies
import numpy as np
import pandas as pd
import os

# Import 2016-2019 crime data
crime_2016 = os.path.join("..","Resources", "crime_clean_2016.csv") 
crime_2017 = os.path.join("..","Resources", "crime_clean_2017.csv")
crime_2018 = os.path.join("..","Resources", "crime_clean_2018.csv")
crime_2019 = os.path.join("..","Resources", "crime_clean_2019.csv")

crime_2016_df_final = pd.read_csv(crime_2016)
crime_2017_df_final = pd.read_csv(crime_2017)
crime_2018_df_final = pd.read_csv(crime_2018)
crime_2019_df_final = pd.read_csv(crime_2019)

# Join datasets for 2016, 2017, 2018 and 2019 for the analysis
join1 = crime_2016_df_final.append(crime_2017_df_final)
join2 = join1.append(crime_2018_df_final)
crime_data = join2.append(crime_2019_df_final)
```

```python
arrests = crime_data[['primary_type','arrest']]
arrests['count'] = 1
```

```python
arrest_count = arrests.groupby(['primary_type','arrest']).agg({'count': 'sum'})
```

```python
arrest_count
```

```python
# reference: https://stackoverflow.com/questions/23377108/pandas-percentage-of-total-with-groupby
arrest_pct = arrest_count.groupby(level=0).apply(lambda x:100 * x / float(x.sum()))
```

```python
arrest_pct= arrest_pct.reset_index()
```

```python
arrest_pct.dtypes
```

```python
solved = arrest_pct[arrest_pct['arrest']==True].sort_values("count",ascending=False)
```
