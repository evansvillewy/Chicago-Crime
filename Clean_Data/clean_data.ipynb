```python
# Dependencies
import json
import requests
import pandas as pd
import datetime as dt
from config import app_token
```

```python
# get the Chicago crime data for 2016-2017
url_crime_2016 = f'https://data.cityofchicago.org/resource/kf95-mnd6.json?$$app_token={app_token}&$limit=269258'
url_crime_2017 = f'https://data.cityofchicago.org/resource/d62x-nvdr.json?$$app_token={app_token}&$limit=268561'
url_crime_2018 = f'https://data.cityofchicago.org/resource/3i3m-jwuy.json?$$app_token={app_token}&$limit=267990'
url_crime_2019 = f'https://data.cityofchicago.org/resource/w98m-zvie.json?$$app_token={app_token}&$limit=259009'
```

```python
#get the json for 2016-2019
response_2016 = requests.get(url_crime_2016).json()
response_2017 = requests.get(url_crime_2017).json()
response_2018 = requests.get(url_crime_2018).json()
response_2019 = requests.get(url_crime_2019).json()
```

```python
#create dataframes for 2016-2019
crime_2016_df=pd.DataFrame(response_2016)
crime_2017_df=pd.DataFrame(response_2017)
crime_2018_df=pd.DataFrame(response_2018)
crime_2019_df=pd.DataFrame(response_2019)
```
