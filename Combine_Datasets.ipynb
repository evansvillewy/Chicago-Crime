```python
# Import Dependencies
import numpy as np
import pandas as pd
import os
```

## Import and Check Chicago Crime Datasets

```python
# Import 2016 crime data
crime_2016 = os.path.join("Resources", "crime_clean_2016.csv") 

crime_2016_df_final = pd.read_csv(crime_2016, index_col='id')

crime_2016_df_final.head()
```

```python
# Import 2017 crime data
crime_2017 = os.path.join("Resources", "crime_clean_2017.csv") 

crime_2017_df_final = pd.read_csv(crime_2017, index_col='id')

crime_2017_df_final.head()
```

```python
# Import 2018 crime data
crime_2018 = os.path.join("Resources", "crime_clean_2018.csv") 

crime_2018_df_final = pd.read_csv(crime_2018, index_col='id')

crime_2018_df_final.head()
```

```python
# Check if columns in dataframes are identical 
if crime_2016_df_final.columns.all() == crime_2017_df_final.columns.all() == crime_2018_df_final.columns.all():
    print("Columns are identical")
else:
    print("Columns are not identical")
```

```python
# Join datasets for 2016, 2017, and 2018
join1 = crime_2016_df_final.append(crime_2017_df_final)
training_df = join1.append(crime_2018_df_final)
training_df
```

```python
# Check sum of lengths of individual datasets match combined length
length1 = len(crime_2016_df_final) + len(crime_2017_df_final) + len(crime_2018_df_final)
if length1 == len(training_df):
    print("Same length")
else: 
    print("Different length")
```
